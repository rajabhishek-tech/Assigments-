```python
df=pd.get_dummies(book)
df.head()
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
