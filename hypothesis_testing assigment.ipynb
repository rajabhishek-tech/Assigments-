#  1)  A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


```python
import pandas as pd
import scipy as sp
import numpy as np
from scipy import stats
import seaborn as sns 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/RajabhishekAditya/Downloads/hypothesis asiigment/Cutlets.csv")
data
```

```python
data.describe()
```

```python
sns.displot(data['Unit A'])
sns.displot(data['Unit B'])
plt.legend(['Unit A', '\n', 'Unit B'])
```

```python
sns.distplot(data['Unit A'])
sns.distplot(data['Unit B'])
plt.legend(['Unit A', 'Unit B'])
```

```python
sns.boxplot(data=[data['Unit A'], data['Unit B']],notch=True)
plt.legend(['Unit A','Unit B'])
```

```python
a=0.05
df1=pd.DataFrame(data['Unit A'])
df2=pd.DataFrame(data['Unit B'])
print(df1,df2)
```

```python
tstat,pValue =sp.stats.ttest_ind(df1,df2) #.f_oneway(when the samples are more than 2 )
print(tstat,pValue)
```

```python
if pValue < 0.05:
    print('accept alternative hypothesis')
else:
    print('accept null hypothesis')
```

# 2)  A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

```python
tstat1,pValue1 =sp.stats.f_oneway(df1,df2,df3,df4) #.f_oneway(when the samples are more than 2 )
print(tstat1,pValue1)
```

```python
if pValue1 < 0.05:
    print('accept alternative hypothesis')
else:
    print('accept null hypothesis')
```

# 5)  Fantaloons Sales managers commented that % of males versus females walking in to the store differ based on day of the week. Analyze the data and determine whether there is evidence at 5 % significance level to support this hypothesis.


```python
import pandas as pd
import scipy as sp
import numpy as np
from scipy import stats
import seaborn as sns 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
Falt_data = pd.read_csv("C:/Users/RajabhishekAditya/Downloads/hypothesis asiigment/Faltoons.csv")
Falt_data
```

```python
Falt_data.describe()
```

```python
a=0.05
df1=pd.DataFrame(Falt_data['Weekdays'])
df2=pd.DataFrame(Falt_data['Weekend'])
print(df1,df2)
```

```python
wdays = Falt_data.Weekdays.value_counts()
print(wdays)
```

```python
wkdays = Falt_data.Weekend.value_counts()
print(wkdays)
```

```python
tstat5,pValue5 =sp.stats.ttest_ind(wdays,wkdays) #.f_oneway(when the samples are more than 2 )
print(tstat5,pValue5)
```

```python
if pValue5 < 0.05:
    print('H0 = there is no difference between male and female customers')
else:
    print('Ha = there is difference between male and female customers ')
```
