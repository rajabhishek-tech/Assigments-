# Perform Clustering for the crime data and identify the number of clusters formed and draw inferences.


```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
Univ = pd.read_csv("C:/Users/RajabhishekAditya/Downloads/clustering assegment/crime_data.csv")
Univ.head(20)
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:])
df_norm.head(20)
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
Univ['h_clusterid'] =y_hc
```

```python
Univ
```

K-Means finding value 


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
crime = pd.read_csv("C:/Users/RajabhishekAditya/Downloads/clustering assegment/crime_data.csv")
crime.head(10)
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(Univ.iloc[:,1:])
scaled_crime_df
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
crime['clusterid_new'] = clusters_new.labels_
crime.head(10)
```

```python
#these are standardized values.
clusters_new.cluster_centers_
```

```python
crime.groupby('clusterid_new').agg(['mean']).reset_index()
```
